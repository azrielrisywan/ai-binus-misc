Praktikum Vicon 2.1 (Artificial Intelligence)

```python
import nltk
nltk.download('movie_reviews')
from nltk.corpus import movie_reviews
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import random

nltk.download('punkt')
nltk.download('stopwords')

# Load sample review
docs = [(list(movie_reviews.words(fileid)), category)
        for category in movie_reviews.categories()
        for fileid in movie_reviews.fileids(category)]

random.shuffle(docs)
print("Sample review:", " ".join(docs[0][0][:50]), "\nSentiment:",
docs[0][1])

def preprocess_words(words):
    stop_words = set(stopwords.words("english"))
    return [w.lower() for w in words if w.isalpha() and w.lower() not in
stop_words]

# Coba untuk 1 review
sample_words = preprocess_words(docs[0][0])
print(sample_words[:50])

all_words = [w.lower() for w in movie_reviews.words()]
all_words = FreqDist(all_words)
word_features = list(all_words)[:2000]

def document_features(document):
    words = set(document)
    features = {}
    for word in word_features:
        features[f'contains({word})'] = (word in words)
    return features

# Training
featuresets = [(document_features(d), c) for (d, c) in docs]
train_set, test_set = featuresets[100:], featuresets[:100]

classifier = nltk.NaiveBayesClassifier.train(train_set)
print("Akurasi:", nltk.classify.accuracy(classifier, test_set))
classifier.show_most_informative_features(10)
```

```python
import spacy
from datasets import load_dataset
from collections import Counter

# Load model multilingual SpaCy
nlp = spacy.load("xx_ent_wiki_sm")

# Load dataset dari Hugging Face
dataset = load_dataset("SEACrowd/indonesian_news_dataset", split="train")

# Tampilkan satu contoh data
print("Contoh Berita:\n")
print("Label:", dataset[0]['label'])
print("Isi Berita:\n", dataset[0]['news'])

# Proses Named Entity Recognition
print("\n=== Named Entity Recognition ===")
sample_text = dataset[0]['news']
doc = nlp(sample_text)

# Tampilkan entitas yang dikenali
print("\nEntitas yang Terdeteksi:")
for ent in doc.ents:
    print(ent.text, "-", ent.label_)

# Analisis entitas dari 10 berita pertama
print("\n=== Ringkasan Entitas dari 10 Berita ===")
all_ents = []

for i in range(10):
    doc = nlp(dataset[i]['news'])
    all_ents.extend([ent.text for ent in doc.ents if ent.label_ != ""])

# Hitung entitas paling sering muncul
ent_counter = Counter(all_ents)
print("\nTop 10 Entitas Paling Sering Muncul:")
for ent, freq in ent_counter.most_common(10):
    print(ent, "-", freq)
```
